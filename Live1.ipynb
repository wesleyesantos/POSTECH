```python
#Bibliotecas principais para utilização em data analytics no python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/content/StudentsPerformance.csv', sep = ';')
df
```

```python
df.info()
#int64 inteiro 64bits
#null nulo
```

```python
#saber quantos comandos nulos tem no dataframe
df.isnull().sum()
```

```python
df.describe().T
#retorna um modelo estatístico
# .T dá um transpose na tabela invertendo-a
# Count conta quantas linhas tem pulando o que é nulo
# Mean média aritmética soma tudo e divide pela quantidade
# std é o desvio padrão, quanto seu dado está distante da média, pra mais ou para menos
# min e max - normal o menor e o maior
# 25% | 50% | 75%  que representam 1o, 2o e 3o quartil
```

```python
sns.boxplot(y = df['math score'])
#O gráfico vem representando o primeiro, segundo e terceiro quartil
#Outlier o que não é comum, são pontos fora da curva, que não se
  #enquadrão no desvio padrão da média em questão
```

```python
df_corr = df[['math score', 'reading score', 'writing score']].corr()
#relação entre tabelas
```

```python
sns.heatmap(df_corr, annot = True, fmt = '.2f')
#mapa de calor
```

```python
dis = df[['math score', 'reading score', 'writing score']].corr()
sns.pairplot(dis)
#Gráfico de histograma e dispersão
#regressão linear vem pra ajustar uma correlação,
  #método dos meninos quadrados será utilizado na aula
```
